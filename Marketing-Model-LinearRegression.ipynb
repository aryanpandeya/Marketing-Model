```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
dataset = pd.read_csv('advertising.csv')
```

```python
dataset.head()
```

```python
X = dataset[['TV','Radio','Newspaper']]
y = dataset['Sales']
X = np.array(X)
y = np.array(y)
```

```python
print(X.shape)
print(y.shape)
```

```python
plt.style.use("seaborn")
plt.plot(X[:,0],y,'bo',label = "TV")
plt.plot(X[:,1],y,'ro',label = "Radio")
plt.plot(X[:,2],y,'yo',label = "Newspaper")
plt.xlabel('Investment in thousands')
plt.ylabel('Sales in millions')
plt.legend()
plt.show()
```

```python
plt.style.use('seaborn-colorblind')
plt.plot(X[:,0],y,"bo")
plt.xlabel("Investment on TV advertising")
plt.ylabel("Sales in millions")
plt.show()
```

```python
plt.plot(X[:,1],y,"ro")
plt.xlabel("Investment on radio advertising")
plt.ylabel("Sales in millions")
plt.show()
```

```python
plt.plot(X[:,2],y,"yo")
plt.xlabel("Investment on newspaper advertising")
plt.ylabel("Sales in millions")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1,random_state = 0)
from sklearn.linear_model import LinearRegression as LR
model = LR().fit(X_train,y_train)
```

```python
coeffs = pd.DataFrame(model.coef_,["TV","Radio","Newspaper"],columns = ["Coefficients"])
coeffs
```

```python
y_pred = model.predict(X_test)
```

```python
preds = pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
preds
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
